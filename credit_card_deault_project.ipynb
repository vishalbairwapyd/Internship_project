**Loan Eligibility Prediction using Machine Learning**

The data set consists of 2000 samples from each of two categories. Five variables are:


1.   Income
2.   Age
1.   Loan
2.   Credit Score
1.   Approved































```python
# Step 1: Import Libraries
import pandas as pd
```

```python
# Step 2: Load Dataset
df = pd.read_csv('https://raw.githubusercontent.com/vishalbairwapyd/Internship_project/main/loan_approval.csv')
```

```python
print(df.head())
```

```python
# Step 3: Dataset Information
print("\nDataset Info:")
print(df.info())
```

```python
print("\nDataset Description:")
print(df.describe())
```

```python
print("\nClass Distribution:")
print(df['Approved'].value_counts())
```

```python
# Step 4: Split Features and Target
X = df.drop('Approved', axis=1)
y = df['Approved']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=42)
```

```python
print("\nTrain/Test Shape:")
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
# Step 5: Select and Train Model
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
print("\nModel Coefficients:")
print(model.coef_)
print("Intercept:", model.intercept_)
```

```python
# Step 6: Predictions
y_pred = model.predict(X_test)
print("\nSample Predictions:")
print(y_pred[:20])
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
# Step 7: Model Evaluation
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
print("\nAccuracy Score:")
print(accuracy_score(y_test, y_pred))
```

```python
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```
